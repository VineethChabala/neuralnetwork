```python
import numpy as np
import pandas as pd
```

```python
#here add your huggingface token
# The rest of your code:
from sentence_transformers import SentenceTransformer
```

```python
embeddings = np.load("movie_embeddings.npy")
df = pd.read_csv("cleaned_movies.csv")
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
query = "A horror movie involving possession and exorcism "
query_vector = model.encode(query)
```

```python
from sentence_transformers import util
```

```python
hits = util.semantic_search(query_vector,embeddings,top_k=10)[0]
for hit in hits:
    # print(hit)
    # hit is a dictionary containing the row index ('corpus_id') and the cosine similarity score ('score')
    movie_row_index = hit['corpus_id']
    similarity_score = hit['score']
    
    # Grab the details from our pandas dataframe
    title = df.iloc[movie_row_index]['title']
    plot = str(df.iloc[movie_row_index]['plot'])
    year = int(df.iloc[movie_row_index]['year'])
    
    print(f"[{similarity_score:.3f}] {title} ({year})")
    # Print a snippet of the plot for context
    print(f"      Plot: {plot[:150]}...") 
    print("-" * 75)
```
